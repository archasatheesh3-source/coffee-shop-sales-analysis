```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['money'], kde=True)
plt.title('Distribution of money spent')
plt.xlabel('Money')
plt.ylabel('Freequency')
plt.show()
```

35-40 money incresing

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='coffee_name',palette='viridis')
plt.title('coffee Type popularity')
plt.xlabel('coffee Type')
plt.ylabel('count')
plt.show()
```

americano coffe type

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='hour',palette='coolwarm')
plt.title('Sales by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.show()
```

10-11 customer incresing tym,customer service,available all items

```python
avg_spend=df.groupby('coffee_name')['money'].mean().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data=avg_spend, x='coffee_name', y='money',palette='magma')
plt.title('Average Spend by Coffee TYpe')
plt.xlabel('coffee name')
plt.ylabel('Average Money')
plt.show()
```

demand cappuccino

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes(include=np.number).corr(),annot=True, cmap='crest')
plt.title('Correlation Heatmap')
plt.show()
```

positive corr

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='day', order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],palette='Spectral')
plt.title('Sales by day of Week')
plt.xlabel('day')
plt.ylabel('count')
plt.show()
```

week day tuesday increasing customer

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='day', order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],palette='Spectral',hue="coffee_name")
plt.title('Sales by day of Week')
plt.xlabel('day')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.countplot(data=df, x='hour',palette='coolwarm',hue="coffee_name")
plt.title('Sales by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.show()
```
